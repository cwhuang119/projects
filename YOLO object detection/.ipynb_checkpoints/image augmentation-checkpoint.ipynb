```python
def color_s(img,w_threshold,c_threshold):
    if img.mode !="RGBA":
        img = img.convert("RGBA")
    data = img.getdata()
    newData = []
    for item in data:
#         if abs(item[0]-255) <w_threshold and abs(item[1]-255)<w_threshold and abs(item[2]- 255)<w_threshold:
#             newData.append((255, 255, 255, 0))
        if item==(0,0,0,0):
            newData.append((0, 0, 0, 0))
        else:
            ran = random.randint(0,4)
            
            r1 = random.randint(-c_threshold,c_threshold)
            r2 = random.randint(-c_threshold,c_threshold)
            r3 = random.randint(-c_threshold,c_threshold)
            r4 = random.randint(-c_threshold,c_threshold)
            newData.append((item[0]+r1, item[1]+r2, item[2]+r2, item[3]+r4))
    img.putdata(newData)
    return img
```

```python
def co(save_path,file,idx):
    obj = cv2.imread(file.path)
    obj = cv2.cvtColor(obj,cv2.COLOR_BGR2HSV)
    r1 = random.randint(80,100)/100
    r2 = random.randint(80,100)/100
    obj[:,:,2] = np.round(obj[:,:,2]*r1)
    obj[:,:,1] = np.round(obj[:,:,1]*r2)
    obj = cv2.cvtColor(obj,cv2.COLOR_HSV2BGR)
    cv2.imwrite(save_path+str(idx)+'@'+file.name,obj)
```

```python
def create(obj,bg):

    o_width, o_height = obj.size
    b_width,b_height = bg.size
    
    
#     r1 = random.randint(0,2)
#     if r1 == 0 :
#         obj = obj.transpose(Image.FLIP_LEFT_RIGHT)


    #flip
#     if r1 ==1 :
#         obj = obj.transpose(Image.FLIP_TOP_BOTTOM)
    
    
    o_width, o_height = obj.size
    

    #resize
    resize_rate = random.randint(5,15)/10
    r_w =math.ceil(o_width*resize_rate)
    r_h =math.ceil(o_height*resize_rate)
    obj=obj.resize((r_w,r_h),resample=0)
    
    
    random_crop = random.randint(90,100)/100
    random_crop_i = 1-random_crop
    start_r = random.randint(0,1)
    start_r1 = random.randint(0,1)
    
    if start_r ==0:
        start_x = 0
        end_x = math.ceil(r_w*random_crop)
    else :
        start_x = math.ceil(r_w*random_crop_i)
        end_x = r_w
    if start_r1 ==0:
        start_y =0
        end_y = math.ceil(r_h*random_crop)
    else :
        start_y =math.ceil(r_h*random_crop_i)
        end_y = r_h
    

    obj=obj.crop((start_x,start_y,end_x,end_y))
    
    
    obj_new = Image.new("RGBA",obj.size,(0,0,0,0))
    obj_new.paste(obj,(0,0))
    
    #rotate
    rotate_rate = random.randint(0,45)
    obj_new = obj_new.rotate(rotate_rate)
    obj_w , obj_h = obj.size
    r = random.randint(0,10)
    if r == 0 : 
        obj_new = obj_new.filter(ImageFilter.BLUR)
    
    new = Image.new("RGBA",bg.size,(0,0,0,0))
    new.paste(bg,(0,0))
    
    new_w,new_h = new.size
    x = random.randint(0,math.floor((new_w-obj_w)))
    y = random.randint(0,math.floor((new_h-obj_h)))
    new.paste(obj_new,(x,y),obj_new)


    x1 = x
    x2 = x+obj_w
    y1 = y
    y2 = y+obj_h
    
    
    return new,(x1,x2,y1,y2)
```

```python
import os
import cv2
from lxml import etree
import xml.etree.cElementTree as ET

def write_xml(folder, img, objects, savedir):
    if not os.path.isdir(savedir):
        os.mkdir(savedir)
    fn = img.name.split('.')[0]
    loc = fn.split('@')[-1].split('_')
    print(loc)
    x1 = loc[0]
    y1 = loc[2]
    x2 = loc[1]
    y2 = loc[3]
    image = cv2.imread(img.path)
    height, width, depth = image.shape
    tl = [(0, 0)]
    br = [(height, width)]
    annotation = ET.Element('annotation')
    ET.SubElement(annotation, 'folder').text = folder
    ET.SubElement(annotation, 'filename').text = img.name
    ET.SubElement(annotation, 'segmented').text = '0'
    size = ET.SubElement(annotation, 'size')
    ET.SubElement(size, 'width').text = str(width)
    ET.SubElement(size, 'height').text = str(height)
    ET.SubElement(size, 'depth').text = str(depth)
    for obj, topl, botr in zip(objects, tl, br):
        ob = ET.SubElement(annotation, 'object')
        ET.SubElement(ob, 'name').text = obj
        ET.SubElement(ob, 'pose').text = 'Unspecified'
        ET.SubElement(ob, 'truncated').text = '0'
        ET.SubElement(ob, 'difficult').text = '0'
        bbox = ET.SubElement(ob, 'bndbox')
        ET.SubElement(bbox, 'xmin').text = str(x1)
        ET.SubElement(bbox, 'ymin').text = str(y1)
        ET.SubElement(bbox, 'xmax').text = str(x2)
        ET.SubElement(bbox, 'ymax').text = str(y2)

    xml_str = ET.tostring(annotation)
    root = etree.fromstring(xml_str)
    xml_str = etree.tostring(root, pretty_print=True)
    save_path = os.path.join(savedir, img.name.replace('jpg', 'xml').replace('jpeg','xml'))
    with open(save_path, 'wb') as temp_xml:
        temp_xml.write(xml_str)

# if __name__ =='__main__':
#     imgs =  [im for im in os.scandir('flag')]
#     for img in imgs:
#         folder='flag'
#         objects=['flag']
#         savedir = 'annotations'
#         write_xml(folder, img, objects, savedir)
```

```python
folder = '../train/train_img'
savedir='../train/train_xml'
imgs =  [im for im in os.scandir(folder)]

for i in imgs:
    if i.name.split('@')[0]=='black':
        objects=['black_tea']
    if i.name.split('@')[0]=='milk':
        objects=['milk_tea']
    if i.name.split('@')[0]=='green':
        objects=['green_tea']
    if i.name.split('@')[0]=='mai':
        objects=['mai_tea']
    write_xml(folder, i, objects, savedir)
```
